## &#10687; Outlier Detection 

```python
# Remove Outlier (IOR) Delivery column
# FIND Q1 AND Q3
Q1=df['Delivery time'].quantile(0.25)
Q3=df['Delivery time'].quantile(0.75)
Q1,Q3
```

```python
IQR = Q3 - Q1
lower_limit = Q1 - (1.5*IQR)
upper_limit = Q3 + (1.5*IQR)

# outliers in Delivery time {column}

outlier_df = df[(df['Delivery time'] < lower_limit) | (df['Delivery time']> upper_limit) ]
outlier_df
```

```python
# Data after removing outliers from delivery time {column} 
df =  df[(df['Delivery time'] >= lower_limit) & (df['Delivery time'] <= upper_limit) ]
df
```

```python
df.shape
```

```python
# Remove Outlier (IOR) Price column
# FIND Q1 AND Q3
Q1 = df['Price'].quantile(0.25)
Q3 = df['Price'].quantile(0.75)
Q1 , Q3
```

```python
IQR = Q3 - Q1
lower_limit = Q1 - (1.5*IQR)
upper_limit = Q3 + (1.5*IQR)

# outliers in Price {column}
outlier_df = df[(df['Price'] < lower_limit) | (df['Price']> upper_limit) ]
outlier_df
```

```python
# Data after removing outliers from Price {column} 
df =  df[(df['Price'] >= lower_limit) & (df['Price'] <= upper_limit) ]
df
```

```python
print(df.describe())
```

```python
df.shape
```

```python
# Remove Outlier (IOR) Avg rating column
# FIND Q1 AND Q3
Q1 = df['Avg ratings'].quantile(0.25)
Q3 = df['Avg ratings'].quantile(0.75)
Q1 , Q3

IQR = Q3 - Q1
lower_limit = Q1 - (1.5*IQR)
upper_limit = Q3 + (1.5*IQR)

# Data after removing outliers from Avg Rating {column} 
df =  df[(df['Avg ratings'] >= lower_limit) & (df['Avg ratings'] <= upper_limit) ]
df
```

```python
df.shape
```

```python
# Remove Outlier (IOR) Total ratings column
# FIND Q1 AND Q3
Q1 = df['Total ratings'].quantile(0.25)
Q3 = df['Total ratings'].quantile(0.75)
Q1 , Q3

IQR = Q3 - Q1
lower_limit = Q1 - (1.5*IQR)
upper_limit = Q3 + (1.5*IQR)

# Data after removing outliers from Total ratings {column} 
df =  df[(df['Total ratings'] >= lower_limit) & (df['Total ratings'] <= upper_limit) ]
df
```
