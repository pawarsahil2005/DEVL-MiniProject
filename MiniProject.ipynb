```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.histplot(df['Avg ratings'],bins = 30)
plt.title('Distribution of Rating')
plt.xlabel('Ratings')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Delivery time'], bins=30, kde=False)
plt.title('Distribution on Delivery Time')
plt.xlabel('Delivery time(min)')
plt.ylabel('Frequency')
plt.show()
```

```python

plt.figure(figsize=(10, 6))
sns.countplot(y='City', data=df, order =df['City'].value_counts().index)
plt.title('Number of Restaurants per City')
plt.xlabel('Count')
plt.ylabel('City')
plt.show()
```

```python
sns.countplot(x ="Avg ratings", data= df )
plt.xlim(0,15)
```
