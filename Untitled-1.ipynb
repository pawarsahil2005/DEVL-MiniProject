```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\vaibhav\Downloads\swiggy1 (1).csv")
plt.figure(figsize=(10,6))
city_counts = df['City'].value_counts()
city_counts.plot(marker='o')
plt.title('Number of Restaurants by City')
plt.xlabel('City')
plt.ylabel('Restaurant Count')
plt.grid()

plt.show()
```

```python
%matplotlib inline

plt.title("Avg rating frequency")
plt.xlabel("Avg rating")
plt.ylabel("frequency")

city_count = df['Avg ratings'].value_counts()
city_count.plot(kind='bar')

plt.show()

```

```python
plt.figure(figsize=(10, 6))
city_count = df['Total ratings'].value_counts()
city_count.plot(kind='bar',color='orange',alpha=0.5)
plt.title("Total ratings frequency",fontsize=16, fontweight='bold')
plt.xlabel("Total ratings", fontsize=14)
plt.ylabel("frequency of Total ratings", fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Price'], bins=30, color='orange', alpha=0.7)
plt.title('Price Distribution of Restaurants')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.histplot(df['Delivery time'], bins=30, kde=True, color='purple')
plt.title('Histogram of Delivery Times')
plt.xlabel('Delivery Time (minutes)')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()
```

```python
numerical_data = df[['Price', 'Avg ratings', 'Total ratings', 'Delivery time']]
correlation_matrix = numerical_data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, cmap='coolwarm', annot = True, fmt='.2f', square=True,cbar=True,linecolor='black',linewidths=0.5)
plt.title('Heatmap of Correlation Matrix')
plt.show()
```

```python
food_type_counts = df['City'].value_counts()
food_type_counts.plot(kind='pie', autopct='%1.1f%%')
plt.title('frequency of orders in each city')
plt.show()
```

```python
food_type_counts = df['Total ratings'].value_counts()
plt.figure(figsize=(10,6))
food_type_counts.plot(kind='pie', autopct='%1.1f%%')
plt.title('frequency of Total rating')
plt.show()
```

```python
top_cities = df['City'].value_counts().nlargest(10).index

# Filter the DataFrame for these top cities
filtered_df = df[df['City'].isin(top_cities)]

# Create a boxplot of Delivery Time grouped by City
filtered_df.boxplot(column='Delivery time', by='City')
plt.title('Delivery Time by City (Top 10 Cities)')
plt.xlabel('City')
plt.ylabel('Delivery Time (mins)')
plt.xticks(rotation=45)  # Rotate x labels for better visibility
plt.show()
```
